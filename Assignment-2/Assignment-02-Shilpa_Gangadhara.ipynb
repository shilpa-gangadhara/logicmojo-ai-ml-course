```python
#Create a 10 x 10 arrays of zeros and then "frame" it with a border of ones.

import numpy as np
arr = np.zeros([10, 10])
arr[0, :] = 1
arr[-1, :] = 1
arr[:, 0] = 1
arr[:, -1] = 1

#arr[1, :] = 1
print(arr)

```

```python
#Create an 8 x 8 array with a checkerboard pattern of zeros and ones.

import numpy as np

arr = np.zeros([8,8], dtype=int)

arr[1::2, ::2] = 1 # odd rows, even columns
arr[::2, 1::2] = 1 #even rows, odd columns

print(arr)
```

```python
#Create a function which creates an n*n array with (i,j) -entry equal to i + j

import numpy as np

def create_sum_array(n):
    row_indices = np.arange(n).reshape(n, 1) #this will crete a  column vector  shape(n,1)
    col_indices = np.arange(n).reshape(1, n) #this will crete a  row vector shape(1,n)

    #Broadcasting : column vector  + row vector gives (i+j) for each entry

    return row_indices + col_indices

n = 5
arr = create_sum_array(n)
print(arr)
 
```

```python
#Evaluate cos and sin on the interval [0,1] with 1000 points and then stack the results into a tall array with rows being the 
#cos(x), sin(x))entries.

import numpy as np

x = np.linspace(0,1, 1000)

cos_x = np.cos(x)
sin_x = np.sin(x)

result = np.column_stack((cos_x, sin_x))

print(result)
```

```python
# Create a random (3*5) array using the np.random.rand(3 ,5) function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

import numpy as np

a = np.random.rand(3,5)

sum_all = np.sum(a)

sum_rows = np.sum(a, axis = 1)

sum_cols = np.sum(a, axis = 0)

print(a, sum_all, sum_rows, sum_cols)
```

```python
#Create a random 5*5 array using the function np.random.rand(5, 5). Then sort the rows according to the second column.

import numpy as np

a = np.random.rand(5, 5)

print(a)

sorted_a =  a[a[:1].argsort()]

print(sorted_a)
```
